```python
import math
import numpy as np
from numpy.linalg import norm
```

```python
x1 = int(input("Ingresar coordenada x1 : "))
y1 = int(input("Ingresar coordenada y1 : "))
x2 = int(input("Ingresar coordenada x2 : "))
y2 = int(input("Ingresar coordenada y2 : "))

from scipy.spatial.distance import euclidean

# Función Distancia Manual
def calcular_distancia_euclidiana(x1,y1,x2,y2):
  d=math.sqrt((x2-x1)**2+(y2-y1)**2)
  return round(d,2)

print("--------- Cálculo Distancias --------")
print(f"Punto 1: ({x1},{y1})")
print(f"Punto 2: ({x2},{y2})")
print("--------- Distancia Euclidiana --------")
print("Función Manual:",calcular_distancia_euclidiana(x1,y1,x2,y2))

punto1=[x1,y1]
punto2=[x2,y2]

# Distancia Función Python
euclidean_distancia = euclidean(punto1, punto2)
print("Función Python:",euclidean_distancia)
```

```python
from scipy.spatial.distance import cityblock

# Función Distancia Manual
def calcular_distancia_manhattan(x1,y1,x2,y2):
  dm=abs(x1-x2)+abs(y1-y2)
  return round(dm,2)

print("--------- Distancia Manhattan --------")
print("Función Manual:",calcular_distancia_manhattan(x1,y1,x2,y2))

# Distancia Función Python
Manhattan_distancia = cityblock(punto1, punto2)
print("Función Python:",Manhattan_distancia)
```

```python
from scipy.spatial.distance import hamming
secuencia1 = input("Ingresar Palabra 1: ")
secuencia2 = input("Ingresar Palabra 2: ")

# Función Distancia Manual
def calcular_distancia_hamming(secuencia1,secuencia2):
  c=0
  n=0
  if len(secuencia1) != len(secuencia2):
    print("Las cadenas deben tener la misma longitud")
  else:
    n = [1 for i in range(len(secuencia1)) if secuencia1[i] != secuencia2[i]]
    c = sum(n)
    por=c/len(secuencia1)
    print("Función Manual:",c,"Porcentaje:",por)


print("--------- Distancia Hamming --------")
calcular_distancia_hamming(secuencia1,secuencia2)

# Distancia Función Python
hamming_distancia = hamming(list(secuencia1),list(secuencia2))
print("Función Python:",hamming_distancia)
```

```python
from scipy.spatial.distance import chebyshev

# Función Distancia Manual
def calcular_distancia_chebyshev(x1,y1,x2,y2):
  dcs=max(abs(x1-x2),abs(y1-y2))
  return round(dcs,2)
print("--------- Distancia Chebyshev --------")
print("Función Manual:",calcular_distancia_chebyshev(x1,y1,x2,y2))

vec1=[x1,y1]
vec2=[x2,y2]
# Distancia Función Python
chebyshev_distancia = chebyshev(vec1, vec2)
print("Función Python:",chebyshev_distancia)
```

```python
from scipy.spatial.distance import canberra

vector1 = np.array([1,2,6])
vector2 = np.array([1,7,9])

# Función Distancia Manual
def calcular_distancia_canberra(vector1,vector2):
    num=np.abs(vector1-vector2)
    den=np.abs(vector1)+np.abs(vector2)
    dc=np.sum(num/den)
    return dc

print("--------- Distancia Canberra --------")
print("Función Manual:",calcular_distancia_canberra(vector1,vector2))

# Distancia Función Python
canberra_distancia = canberra(vector1,vector2)
print("Función Python:",canberra_distancia)
```

```python
from scipy.spatial.distance import cosine

vector1 = [1, 2, 3]
vector2 = [4, 5, 6]

# Función Distancia Manual
def calcular_distancia_coseno(vector1,vector2):
    num=np.dot(vector1,vector2)
    den=norm(vector1)*norm(vector2)
    co=np.sum(num/den)
    dco=1-co
    return dco

print("--------- Distancia Coseno --------")
print("Función Manual:",calcular_distancia_coseno(vector1,vector2))

# Distancia Función Python
cosine_distancia = cosine(vector1, vector2)
print("Función Python:",cosine_distancia)
```

```python
from sklearn.metrics import jaccard_score

conjunto1 = [1, 0, 1, 1, 1, 0]
conjunto2 = [0, 1, 1, 0, 1, 0]

# Función Distancia Manual
def calcular_distancia_jacard(conjunto1,conjunto2):
   conjunto_i= []
   tam_i=0
   tam_u=0
   for i in range(min(len(conjunto1), len(conjunto2))):
    if conjunto1[i] == conjunto2[i] and conjunto1[i] ==1:
     tam_i+=1

    if conjunto1[i] == 1 or conjunto2[i] ==1:
     tam_u+=1

   dj=1-(tam_i/tam_u)
   print("Tamaño Unión:",tam_u,"Tamaño Intersección:",tam_i)

   print("Función Manual:",round(dj,2))

print("--------- Distancia Jacard --------")
calcular_distancia_jacard(conjunto1,conjunto2)

# Distancia Función Python
jaccard_distancia = 1 - jaccard_score(conjunto1,conjunto2)
print("Función Python:",jaccard_distancia)
```
