```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Ejemplo de datos
X = [ 1, 2, 3, 4, 5 ]
h = [ -1, 5, 3, -2, 1]

# Definir los indices donde el valor cero esta ubicado
X_zero_index = 1
h_zero_index = 2

# Calcular 1a convolución de X y h
Y = np.convolve(X, h)

# Definir los indices de las secuencias para la grafica
X_indices = np.arange(-X_zero_index, len(X) - X_zero_index)
h_indices = np.arange(-h_zero_index, len(h) - h_zero_index)
Y_indices = np.arange(X_indices[0] + h_indices[0], X_indices[-1] + h_indices[-1] + 1)

# Convolución de x y h
Y = np.convolve(X, h)
```

```python
# Crear la gráfica
plt.figure(figsize=(12, 8))

# Graficar X
plt.subplot(3, 1, 1)
plt.stem(X_indices, X, use_line_collection=True)
plt.title('X')
plt.xlabel('n')
plt.ylabel('X[n]')

# Graficar h
plt.subplot(3, 1, 2)
plt.stem(h_indices, h, use_line_collection=True)
plt.title('h')
plt.xlabel('n')
plt.ylabel('h[n]')

# Graficar Y = conv(X, h)
plt.subplot(3, 1, 3)
plt.stem(Y_indices, Y, use_line_collection=True)
plt.title('Y = conv(X, h)')
plt.xlabel('n')
plt.ylabel('Y[n]')

plt.tight_layout()
plt.show()
```
