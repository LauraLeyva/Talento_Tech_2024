```python
import skimage

# Ruta de la imagen

image_path = 'Tucan.jpg'

# Cargar la imagen

image = skimage.io.imread(image_path)
# Mostrar la imagen
skimage.io.imshow(image)
skimage.io.show()
```

```python
# Rotar la imagen 90 grados en sentido antihorario
rotated_image = skimage.transform.rotate(image, angle=30)
# Mostrar la imagen rotada
skimage.io.imshow(rotated_image)

skimage.io.show()
```

```python
# Aplicar un filtro Gaussiano para suavizar la imagen
smoothed_image = skimage.filters.gaussian(image, sigma=2)

# Mostrar la imagen suavizada
skimage.io.imshow(smoothed_image)

skimage.io.show()
```
