Procesamiento datos

```python
# Preprocesamiento de datos: tokenizacion y extraccion de caracteristicas
def preprocess(text):
  tokens = nltk.word_tokenize(text)
  return {word: True for word in tokens}
```

Aplicación procesamiento de datos

```python
# Aplicamos el preprocesamiento a los datos
featuresets = [(preprocess(text), label) for (text, label) in data]
print("---Feature Set---")
print(featuresets)
```

```python
import matplotlib.pyplot as plt
from collections import Counter
# Extraer todas las palabras
all_words = []
for features, label in featuresets:
    all_words.extend(features.keys())

# Contar la frecuencia de cada palabra
word_counts = Counter(all_words)

# Ordenar las palabras por frecuencia
sorted_word_counts = sorted(word_counts.items(), key=lambda x: x[1], reverse=True)
words, counts = zip(*sorted_word_counts)

# Crear el histograma
plt.figure(figsize=(15, 7))
plt.bar(words, counts, color='blue', edgecolor='black', alpha=0.7)
plt.xlabel('Words')
plt.ylabel('Frequency')
plt.title('Frequency Distribution of Words in Featuresets')
plt.xticks(rotation=90)
plt.grid(axis='y')

# Mostrar el histograma
plt.show()
```

```python
from wordcloud import WordCloud
# Crear un diccionario de frecuencias para wordcloud
word_freq = dict(word_counts)

# Crear la nube de palabras
wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(word_freq)

# Mostrar la nube de palabras
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

División de datos

```python
# Dividimos los datos en conjuntos de entrenamiento y prueba
train_set, test_set = featuresets[:16], featuresets[16:]
print("----Train Set---")
print(train_set)
print("----Test Set---")
print(test_set)
```

Entrenamiento clasificador

```python
# Entrenamos un clasificador utilizando Naive Bayes
classifier = nltk.NaiveBayesClassifier.train(train_set)
```

Evaluación Clasificador

```python
# Evaluamos el clasificador en el conjunto de prueba
accuracy = nltk.classify.accuracy(classifier, test_set)
print("Accuracy:", accuracy)
```

Clasificación Nuevo Texto

```python
# Clasificamos un nuevo texto
new_text = "This movie is amazing"
new_text_features = preprocess(new_text)
predicted_label = classifier.classify(new_text_features )
print("Predicted label 1:", predicted_label)

new_text = "This film is terrible"
new_text_features = preprocess(new_text)
predicted_label = classifier.classify(new_text_features )
print("Predicted label 2:", predicted_label)

new_text = "I like this movie"
new_text_features = preprocess(new_text)
predicted_label = classifier.classify(new_text_features )
print("Predicted label 3:", predicted_label)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, ConfusionMatrixDisplay, roc_curve, roc_auc_score
# Separar características y etiquetas del conjunto de prueba
test_features, test_labels = zip(*test_set)

# Predecir las etiquetas para el conjunto de prueba
predicted_labels = [classifier.classify(features) for features in test_features]

# Calcular las métricas
accuracy = accuracy_score(test_labels, predicted_labels)
precision = precision_score(test_labels, predicted_labels, average='weighted')
recall = recall_score(test_labels, predicted_labels, average='weighted')
f1 = f1_score(test_labels, predicted_labels, average='weighted')

# Imprimir las métricas
print(f'Accuracy: {accuracy:.2f}')
print(f'Precision: {precision:.2f}')
print(f'Recall: {recall:.2f}')
print(f'F1-Score: {f1:.2f}')

# Calcular la matriz de confusión
conf_matrix = confusion_matrix(test_labels, predicted_labels)
print(f'Confusion Matrix:\n{conf_matrix}')

# Mostrar la matriz de confusión
disp = ConfusionMatrixDisplay(confusion_matrix=conf_matrix)
disp.plot(cmap=plt.cm.Blues)
plt.show()
```

```python
from sklearn.metrics import roc_auc_score, roc_curve, auc
# Obtener características y etiquetas del conjunto de prueba
test_features, test_labels = zip(*test_set)

# Predecir las etiquetas para el conjunto de prueba
predicted_labels = [classifier.classify(features) for features in test_features]

# Calcular las probabilidades para el conjunto de prueba
probabilities = [classifier.prob_classify(features).prob('positive') for features in test_features]

# Convertir etiquetas a formato binario para sklearn
test_labels_binary = [1 if label == 'positive' else 0 for label in test_labels]
predicted_labels_binary = [1 if label == 'positive' else 0 for label in predicted_labels]

# Calcular las métricas
accuracy = accuracy_score(test_labels_binary, predicted_labels_binary)
precision = precision_score(test_labels_binary, predicted_labels_binary)
recall = recall_score(test_labels_binary, predicted_labels_binary)
f1 = f1_score(test_labels_binary, predicted_labels_binary)

# Calcular la curva ROC y el AUC
fpr, tpr, _ = roc_curve(test_labels_binary, probabilities)
roc_auc = auc(fpr, tpr)

# Graficar la curva ROC
plt.figure()
plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")
plt.show()
```
