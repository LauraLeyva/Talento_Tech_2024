Ejemplo convolución de una imagen utilizando Opencv en python

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
# Cargar la imagen
image = cv2.imread('tortuga.jpg')

# Definir el kernel de convolución
kernel = np.array([[0, -1, 0],
                   [-1, 5, -1],
                   [0, -1, 0]])

# Aplicar la convolución
convolved_image = cv2.filter2D(image, -1, kernel)
# Mostrar la imagen original y la imagen convolucionada
cv2_imshow(image)
cv2_imshow(convolved_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Definir el kernel de convolución
kernel = np.array([[-1, 0, 1],
                   [-2, 0, 2],
                   [-1, 0, 1]])

# Aplicar la convolución
convolved_image = cv2.filter2D(image, -1, kernel)
# Mostrar la imagen original y la imagen convolucionada
cv2_imshow(image)
cv2_imshow(convolved_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Definir el kernel de convolución
kernel = np.array([[-1, -2, -1],
                   [ 0, 0,  0],
                   [ 1, 2,  1]])

# Aplicar la convolución
convolved_image = cv2.filter2D(image, -1, kernel)
# Mostrar la imagen original y la imagen convolucionada
cv2_imshow(image)
cv2_imshow(convolved_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
