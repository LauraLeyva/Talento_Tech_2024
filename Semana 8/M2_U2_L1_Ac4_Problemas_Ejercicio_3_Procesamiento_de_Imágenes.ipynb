```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
faceClassif = cv2.CascadeClassifier('haarcascade_frontalcatface.xml')
```

```python
# Cargar la imagen
#image = cv2.imread('gatos.jpeg')
image = cv2.imread('gatos2.jpg')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray)
```

```python
#Deteccion de rostros de gatos:
faces = faceClassif.detectMultiScale(gray,
scaleFactor=1.1,
minNeighbors=5,
minSize=(30,30),
maxSize=(200,200))
print (faces)
```

```python
# Dibujar rectangulos alrededor de los rostros detectados
for (x,y,w,h) in faces:
   cv2.rectangle(image, (x,y), (x+w,y+h),(0,255,0),2)
```

```python
# Mostrar la imagen con los contornos resaltados
cv2_imshow(image)
cv2.waitKey (0)
cv2.destroyAllWindows()
```
